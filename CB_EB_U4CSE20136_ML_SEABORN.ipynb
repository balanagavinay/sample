```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data = sns.load_dataset("iris")
```

```python
data.head()
```

```python
sns.get_dataset_names()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# setting the title using Matplotlib
plt.title('Title using Matplotlib Function')

plt.xlim(4, 8)
  
plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# changing the theme to dark (darkgrid/whitegrid/dark/white/ticks)
sns.set_style("darkgrid")
plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# Removing the spines
sns.despine()
plt.show()
```

```python
plt.figure(figsize = (10, 6))
  
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# Removing the spines
sns.despine()
  
plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# Setting the scale of the plot
sns.set_context("talk")
  
plt.show()
```

```python
palette = sns.color_palette()
  
# plots the color palette as a horizontal array
sns.palplot(palette)
  
plt.show()
```

```python
palette = sns.color_palette('coolwarm', 15)
  
# diverging color palette
sns.palplot(palette)
  
plt.show()
```

```python
palette = sns.color_palette('YlGn', 15)
  
# sequential color palette
sns.palplot(palette)
  
plt.show()
```
