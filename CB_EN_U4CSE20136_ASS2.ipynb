```python
import pandas as pd
df=pd.read_csv("https://github.com/flaviovdf/evcomp2018/raw/master/data/educ_inc.csv")
df
```

```python
df.groupby(df['Educational Attainment'])['Population Count'].sum()
```

```python
df1=df.groupby(['Age','Educational Attainment'])['Population Count'].sum()
df3 = df1.groupby(level=0).apply(lambda x:100 * x / float(x.sum()))
df3
```
